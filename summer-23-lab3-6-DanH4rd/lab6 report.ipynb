# Porównywanie tuningów

| Methoda | accuracy | czas (sekundy) |
| ------ | -------- | --------------- |
| RandomSearh | 0.7267 | 584.89 |
| HalvingRandomSearch | 0.6631 | 256.82 |
| GridSearch | 0.7263 | 665.15 |
| HalvingGridSearch | 0.6676 | 372.59 |

```python
import matplotlib.pyplot as plt

# define method names and corresponding accuracy scores
methods = ['Random Search', 'Halving Random Search', 'Grid Search', 'Halving Grid Search']
scores = [0.7267, 0.6631, 0.7263, 0.6676]

# create bar plot
fig, ax = plt.subplots()

ax.bar(methods, scores)
fig.set_size_inches(7, 5)

# add labels and title
ax.set_xlabel('Hyperparam Tuning Method')
ax.set_ylabel('Accuracy Score')
ax.set_title('Comparison of Feature Selection Methods')
ax.set_ylim(bottom=0.6)
# display plot
plt.tight_layout()
plt.show()
```
